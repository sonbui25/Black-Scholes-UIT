```python
import numpy as np
import matplotlib.pyplot as plt
```

### Ta sẽ cho các thành phần sau cố định

$\text{Giá cổ phiếu ban đầu}: \quad S_0 = 100$

$\text{Khoảng thời gian (tính theo năm)}: \quad T = 1.0$

$\text{Số bước thời gian}: \quad N = 300$

$\text{Số đường mô phỏng}: \quad \text{num_paths} = 15$

### Ví dụ 1 - Tỷ suất lợi nhuận dương | Phương sai nhỏ

$\text{Tỷ suất sinh lời kỳ vọng}: \quad \mu = 0.1$

$\text{Độ biến động}: \quad \sigma = 0.1$


```python
# Parameters
np.random.seed(42)  # For reproducibility

S0 = 100        
mu = 0.1     
sigma = 0.1      
T = 1.0          
N = 300          
num_paths = 15     

# Time increment
dt = T / N
t = np.linspace(0, T, N)

# Generate paths
paths = np.zeros((num_paths, N))
for i in range(num_paths):
    Wt = np.random.standard_normal(size=N) * np.sqrt(dt)  # Brownian increments
    Wt = np.cumsum(Wt)  # Cumulative sum to get Brownian path
    paths[i] = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * Wt)

# Plotting the results
plt.figure(figsize=(12, 8))
for i in range(num_paths):
    plt.plot(t, paths[i], lw=1.5, label=f'Path {i+1}')

plt.title('Chuyển Động Brownian Hình Học trong trường hợp Tỷ suất lợi nhuận dương | Phương sai nhỏ')
plt.xlabel('Time (Years)')
plt.ylabel('Stock Price')
# plt.legend()
plt.grid(True)
plt.show()
```

### Ví dụ 2 - Tỷ suất lợi nhuận âm | Phương sai nhỏ

$\text{Tỷ suất sinh lời kỳ vọng}: \quad \mu = -0.1$

$\text{Độ biến động}: \quad \sigma = 0.1$


```python
# Parameters
np.random.seed(42)  # For reproducibility

S0 = 100        
mu = -0.1     
sigma = 0.1      
T = 1.0          
N = 300          
num_paths = 15     

# Time increment
dt = T / N
t = np.linspace(0, T, N)

# Generate paths
paths = np.zeros((num_paths, N))
for i in range(num_paths):
    Wt = np.random.standard_normal(size=N) * np.sqrt(dt)  # Brownian increments
    Wt = np.cumsum(Wt)  # Cumulative sum to get Brownian path
    paths[i] = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * Wt)

# Plotting the results
plt.figure(figsize=(12, 8))
for i in range(num_paths):
    plt.plot(t, paths[i], lw=1.5, label=f'Path {i+1}')

plt.title('Chuyển Động Brownian Hình Học trong trường hợp Tỷ suất lợi nhuận âm | Phương sai nhỏ')
plt.xlabel('Time (Years)')
plt.ylabel('Stock Price')
# plt.legend()
plt.grid(True)
plt.show()
```

Ta thấy ở cả 2 ví dụ khi $\mu$ (độ trôi của $\textbf{GBM}$) $> 0$  hoặc $< 0$ và $\sigma$ (độ biến động của yếu tố ngẫu nhiên) nhỏ $< 15\%$, thì đa số đường minh họa có xu hướng cùng hội tụ ở một vùng lân cận.
